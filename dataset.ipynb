```python
data = pd.read_csv('./input/flights.csv')
```

```python
date_counts = data['date'].value_counts().sort_index()

print("\n날짜별 데이터 개수:")
print(date_counts)

# 날짜별 데이터 개수 시각화
plt.figure(figsize=(12, 6))
sns.barplot(x=date_counts.index, y=date_counts.values)
plt.title('Number of Data Points per Date')
plt.xlabel('Date')
plt.ylabel('Number of Data Points')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 데이터 개수의 기술 통계량 계산
count_stats = date_counts.describe()

print("\n날짜별 데이터 개수의 기술 통계량:")
print(count_stats)

# 날짜별 데이터 개수의 히스토그램
plt.figure(figsize=(10, 6))
sns.histplot(date_counts.values, kde=True)
plt.title('Distribution of Data Points per Date')
plt.xlabel('Number of Data Points')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 가장 많은 데이터를 가진 상위 5개 날짜
top_5_dates = date_counts.nlargest(5)
print("\n데이터가 가장 많은 상위 5개 날짜:")
print(top_5_dates)

# 가장 적은 데이터를 가진 하위 5개 날짜
bottom_5_dates = date_counts.nsmallest(5)
print("\n데이터가 가장 적은 하위 5개 날짜:")
print(bottom_5_dates)
```
