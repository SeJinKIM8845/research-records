```python
#!/usr/bin/env python3
import sys
import os
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from filter import DroneDataFilter
import warnings
import json

# 경고 메시지 필터링
warnings.filterwarnings('ignore')

def analyze_raw_data(filter_obj):
    print("Step 1: Loading and analyzing raw data...")
    raw_data = {}
    
    # Primary sensors
    print("\nLoading primary sensors data...")
    for sensor_type in filter_obj.config['sensors']['primary'].keys():
        print(f"\nAnalyzing {sensor_type} data:")
        data = filter_obj.get_sensor_data(sensor_type, 'primary')
        if not data.empty:
            raw_data[sensor_type] = data
            print(f"Shape: {data.shape}")
            
            # 시간 범위 확인
            if '_time' in data.columns:
                print(f"Time range: {data['_time'].min()} to {data['_time'].max()}")
            
            # 시뮬레이션 ID 확인
            if 'simulation_id' in data.columns:
                print(f"Number of unique simulation IDs: {data['simulation_id'].nunique()}")
                print("\nSimulation ID distribution:")
                print(data['simulation_id'].value_counts())
            
            # 수치형 데이터 통계
            numeric_cols = data.select_dtypes(include=[np.number]).columns
            if len(numeric_cols) > 0:
                print("\nNumerical Features Statistics:")
                print(data[numeric_cols].describe())
            
            # 결측치 확인
            missing_values = data.isnull().sum()
            if missing_values.any():
                print("\nMissing Values:")
                print(missing_values[missing_values > 0])
    
    # Secondary sensors
    print("\nLoading secondary sensors data...")
    for sensor_type in filter_obj.config['sensors']['secondary'].keys():
        print(f"\nAnalyzing {sensor_type} data:")
        data = filter_obj.get_sensor_data(sensor_type, 'secondary')
        if not data.empty:
            raw_data[sensor_type] = data
            # ... 동일한 분석 수행
    
    return raw_data

def visualize_data(raw_data):
    print("\nStep 2: Data Visualization")
    for sensor_type, data in raw_data.items():
        print(f"\nVisualizing {sensor_type} data:")
        numeric_cols = data.select_dtypes(include=[np.number]).columns
        
        if len(numeric_cols) > 0:
            # 시계열 플롯
            plt.figure(figsize=(15, 5*len(numeric_cols[:3])))
            for i, col in enumerate(numeric_cols[:3], 1):
                plt.subplot(len(numeric_cols[:3]), 1, i)
                plt.plot(data[col][:1000])  # 처음 1000개 샘플만
                plt.title(f'{sensor_type} - {col}')
                plt.grid(True)
            plt.tight_layout()
            plt.show()

def main():
    try:
        # Filter 인스턴스 생성
        filter_obj = DroneDataFilter(config_path='filter_config.yaml')
        
        # 1. 데이터 로드 및 분석
        raw_data = analyze_raw_data(filter_obj)
        input("Press Enter to continue to visualization...")
        
        # 2. 데이터 시각화
        visualize_data(raw_data)
        input("Press Enter to continue to processing...")
        
        # 3. 데이터 전처리
        print("\nStep 3: Data Processing")
        sequences, scaler_params = filter_obj.filter_batch(raw_data)
        
        # 4. 전처리 결과 출력
        print("\nProcessing Results:")
        print(f"Sequences shape: {sequences.shape}")
        print(f"Number of sequences: {len(sequences)}")
        print(f"Sequence length: {sequences.shape[1]}")
        print(f"Number of features: {sequences.shape[2]}")
        
        # 5. 저장된 파일 확인
        print("\nSaved Files:")
        save_path = Path(filter_obj.config['paths']['base_dir'])
        for file in save_path.glob('*'):
            print(f"\n{file.name}")
            
        filter_obj.cleanup()
        
    except Exception as e:
        print(f"Error occurred: {str(e)}")
        import traceback
        print(traceback.format_exc())

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

def analyze_sequences():
    """시퀀스 데이터 분석"""
    # 데이터 로드
    sequences = np.load('processed_data/batch_sequences.npy')
    
    # 기본 통계
    stats = {
        'shape': sequences.shape,
        'mean': np.mean(sequences),
        'std': np.std(sequences),
        'min': np.min(sequences),
        'max': np.max(sequences)
    }
    print("\n데이터 기본 통계:")
    for k, v in stats.items():
        print(f"{k}: {v}")
    
    # 특성별 분포
    feature_stats = {
        'means': np.mean(sequences, axis=(0,1)),
        'stds': np.std(sequences, axis=(0,1)),
        'mins': np.min(sequences, axis=(0,1)),
        'maxs': np.max(sequences, axis=(0,1))
    }
    
    print("\n특성별 통계:")
    for i in range(sequences.shape[2]):
        print(f"\nFeature {i}:")
        print(f"Mean: {feature_stats['means'][i]:.4f}")
        print(f"Std: {feature_stats['stds'][i]:.4f}")
        print(f"Min: {feature_stats['mins'][i]:.4f}")
        print(f"Max: {feature_stats['maxs'][i]:.4f}")
    
    # 시각화
    plt.figure(figsize=(15, 10))
    
    # 전체 데이터 분포
    plt.subplot(2, 1, 1)
    plt.hist(sequences.flatten(), bins=50)
    plt.title('Overall Data Distribution')
    
    # 특성별 분포
    plt.subplot(2, 1, 2)
    plt.boxplot([sequences[:,:,i].flatten() for i in range(sequences.shape[2])])
    plt.title('Feature-wise Distribution')
    plt.xticks(range(1, sequences.shape[2]+1), [f'F{i}' for i in range(sequences.shape[2])])
    
    plt.tight_layout()
    plt.savefig('data_analysis.png')
    plt.close()

def verify_sequences():
    """시퀀스 연속성 검증"""
    sequences = np.load('processed_data/batch_sequences.npy')
    
    # 연속성 체크
    discontinuities = []
    for i in range(len(sequences)):
        seq = sequences[i]
        diffs = np.abs(np.diff(seq, axis=0))
        max_diff = np.max(diffs)
        if max_diff > 0.5:  # 임계값
            discontinuities.append((i, max_diff))
    
    print(f"\n큰 변화가 있는 시퀀스 수: {len(discontinuities)}")
    if discontinuities:
        print("상위 10개 불연속 지점:")
        for idx, diff in sorted(discontinuities, key=lambda x: x[1], reverse=True)[:10]:
            print(f"Sequence {idx}: Max difference {diff:.4f}")

if __name__ == "__main__":
    analyze_sequences()
    verify_sequences()
```
