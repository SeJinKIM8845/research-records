```python
data=pd.read_csv('1.14 사고.log_unified.csv')
data.head(100) # 데이터프레임 상위 5개 행을 반환
```

```python
import matplotlib.pyplot as plt

# GPS 신호 이상 구간 데이터 추출 (340초 ~ 380초로 가정)
gps_anomalous_data = data[(data['timestamp'] >= 300) & (data['timestamp'] <= 350)]

# 진동 데이터와 클리핑 데이터 시각화
def plot_vibration_and_clipping(data):
    fig, ax1 = plt.subplots(figsize=(14, 8))

    # 진동 데이터 시각화 (VibeX, VibeY, VibeZ)
    ax1.plot(data['timestamp'], data['VibeX'], label='VibeX', color='blue', linewidth=2)
    ax1.plot(data['timestamp'], data['VibeY'], label='VibeY', color='orange', linestyle='--', linewidth=2)
    ax1.plot(data['timestamp'], data['VibeZ'], label='VibeZ', color='green', linestyle='-.', linewidth=2)
    ax1.set_xlabel('Timestamp (s)', fontsize=14)
    ax1.set_ylabel('Vibration Levels', fontsize=14)
    ax1.tick_params(axis='y')
    ax1.legend(loc='upper left', fontsize=12)
    ax1.set_title('Vibration and Clipping During GPS Signal Loss (300s - 350s)', fontsize=16)

    # 클리핑 데이터 시각화 (Clip)
    ax2 = ax1.twinx()  # 공유 x축
    ax2.plot(data['timestamp'], data['Clip'], label='Clip', color='red', linestyle=':', linewidth=2)
    ax2.set_ylabel('Clip Value', fontsize=14)
    ax2.tick_params(axis='y', labelcolor='red')
    ax2.legend(loc='upper right', fontsize=12)

    # 그래프 출력
    fig.tight_layout()
    plt.grid()
    plt.show()

# 함수 호출: 진동 및 클리핑 데이터 시각화
plot_vibration_and_clipping(gps_anomalous_data)
```

```python
# 라이브러리 임포트
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 데이터 로드 및 필터링
file_path = '1.14 사고.log_unified.csv'  # 데이터 파일 경로
df = pd.read_csv(file_path)

# datetime 컬럼을 datetime 타입으로 변환
df['datetime'] = pd.to_datetime(df['datetime'])

# 300초 ~ 350초 데이터 필터링
filtered_df = df[(df['timestamp'] >= 300) & (df['timestamp'] <= 350)]

# 시각화 스타일 설정
sns.set_theme(style="whitegrid", palette="muted")

# x축, y축 선 추가를 위한 함수 정의
def add_axis_lines():
    plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # y=0 축
    plt.axvline(filtered_df['timestamp'].min(), color='black', linewidth=0.8, linestyle='--')  # x 최소
    plt.axvline(filtered_df['timestamp'].max(), color='black', linewidth=0.8, linestyle='--')  # x 최대

# 2. Mode 변경 시각화
mode_changes = filtered_df['Mode'].shift() != filtered_df['Mode']  # Mode 변경 탐지
plt.figure(figsize=(10, 6))
sns.scatterplot(x='timestamp', y='Mode', data=filtered_df, hue='Mode', palette='tab10', s=100)
add_axis_lines()
plt.xlabel('Timestamp (seconds)', fontsize=12)
plt.ylabel('Mode', fontsize=12)
plt.title('Mode Changes (300s - 350s)', fontsize=16)
plt.legend(title='Mode', loc='upper right')
plt.grid()
plt.show()

# Mode 변경된 시점 출력
print("Mode 변경된 시점:")
print(filtered_df[mode_changes][['timestamp', 'Mode']])

# 3. Subsys 및 ECode 데이터 시각화
plt.figure(figsize=(10, 6))
sns.scatterplot(x='timestamp', y='Subsys', data=filtered_df, label='Subsystem ID', color='blue', s=50)
sns.lineplot(x='timestamp', y='Subsys', data=filtered_df, label='Subsystem Trend', color='blue')
plt.scatter(filtered_df['timestamp'], filtered_df['ECode'], label='ECode', color='orange', s=50)
add_axis_lines()
plt.xlabel('Timestamp (seconds)', fontsize=12)
plt.ylabel('Subsystems and Error Codes', fontsize=12)
plt.title('Subsystem and ECode Data (300s - 350s)', fontsize=16)
plt.legend()
plt.grid()
plt.show()
```
