```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 로드
data = pd.read_csv("1.14 사고.log_unified.csv")  # 데이터 파일명을 설정하세요.

# 데이터 확인
print(data.info())  # 데이터 타입과 결측값 확인
print(data.describe())  # 기본 통계량 확인

# 시각화를 위한 설정
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# 1. 시간에 따른 주요 값 변화 시각화
def plot_time_series(data, column, title, ylabel):
    plt.figure()
    plt.plot(data['timestamp'], data[column], label=column)
    plt.title(title)
    plt.xlabel('Timestamp (s)')
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid()
    plt.show()

# 주요 값들 시각화
plot_time_series(data, 'gps_altitude', 'GPS Altitude Over Time', 'GPS Altitude (m)')
plot_time_series(data, 'ground_speed', 'Ground Speed Over Time', 'Ground Speed (m/s)')
plot_time_series(data, 'Roll', 'Roll Over Time', 'Roll (degrees)')
plot_time_series(data, 'Pitch', 'Pitch Over Time', 'Pitch (degrees)')
plot_time_series(data, 'Yaw', 'Yaw Over Time', 'Yaw (degrees)')
plot_time_series(data, 'baro_alt', 'Barometric Altitude Over Time', 'Barometric Altitude (m)')

# 2. 모터 데이터 시각화
plt.figure()
for motor in ['motor_1', 'motor_2', 'motor_3', 'motor_4']:
    plt.plot(data['timestamp'], data[motor], label=motor)
plt.title('Motor Performance Over Time')
plt.xlabel('Timestamp (s)')
plt.ylabel('Motor Values')
plt.legend()
plt.grid()
plt.show()

# 3. 진동 데이터 시각화
plt.figure()
for vibe in ['VibeX', 'VibeY', 'VibeZ']:
    plt.plot(data['timestamp'], data[vibe], label=vibe)
plt.title('Vibration Data Over Time')
plt.xlabel('Timestamp (s)')
plt.ylabel('Vibration Levels')
plt.legend()
plt.grid()
plt.show()

# 4. 상관 관계 히트맵
correlation_matrix = data[['gps_altitude', 'ground_speed', 'Roll', 'Pitch', 'Yaw', 
                           'baro_alt', 'VibeX', 'VibeY', 'VibeZ', 'motor_1', 
                           'motor_2', 'motor_3', 'motor_4']].corr()
plt.figure()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# 5. 클리핑 데이터 강조
clipped_data = data[data['Clip'] != 0]
if not clipped_data.empty:
    print("Clipping occurred at the following timestamps:")
    print(clipped_data[['timestamp', 'Clip']])
    plt.figure()
    plt.scatter(clipped_data['timestamp'], clipped_data['Clip'], color='red', label='Clipping')
    plt.title('Clipping Events')
    plt.xlabel('Timestamp (s)')
    plt.ylabel('Clip Value')
    plt.legend()
    plt.grid()
    plt.show()
else:
    print("No clipping events detected.")
```

```python
import matplotlib.pyplot as plt

# 이상 구간 데이터 추출
anomalous_data = data[(data['timestamp'] >= 300) & (data['timestamp'] <= 400)]

# GPS 데이터 시각화 함수
def plot_gps_and_altitude(data):
    plt.figure(figsize=(14, 8))
    
    # GPS 고도 시각화
    plt.plot(data['timestamp'], data['gps_altitude'], label='GPS Altitude (m)', color='blue', linewidth=2)
    
    # 기압 고도 시각화
    plt.plot(data['timestamp'], data['baro_alt'], label='Barometric Altitude (m)', color='orange', linestyle='--', linewidth=2)
    
    # 그래프 레이블 및 제목
    plt.title('Comparison of GPS and Barometric Altitude (300s - 400s)', fontsize=16)
    plt.xlabel('Timestamp (s)', fontsize=14)
    plt.ylabel('Altitude (m)', fontsize=14)
    plt.legend(fontsize=12)
    plt.grid(True)
    
    # 시각화 출력
    plt.show()

# GPS 시각화 호출
plot_gps_and_altitude(anomalous_data)
```

```python
import matplotlib.pyplot as plt

# GPS 신호 이상 구간 데이터 추출 (340초 ~ 380초로 가정)
gps_anomalous_data = data[(data['timestamp'] >= 240) & (data['timestamp'] <= 380)]

# 진동 데이터와 클리핑 데이터 시각화
def plot_vibration_and_clipping(data):
    fig, ax1 = plt.subplots(figsize=(14, 8))

    # 진동 데이터 시각화 (VibeX, VibeY, VibeZ)
    ax1.plot(data['timestamp'], data['VibeX'], label='VibeX', color='blue', linewidth=2)
    ax1.plot(data['timestamp'], data['VibeY'], label='VibeY', color='orange', linestyle='--', linewidth=2)
    ax1.plot(data['timestamp'], data['VibeZ'], label='VibeZ', color='green', linestyle='-.', linewidth=2)
    ax1.set_xlabel('Timestamp (s)', fontsize=14)
    ax1.set_ylabel('Vibration Levels', fontsize=14)
    ax1.tick_params(axis='y')
    ax1.legend(loc='upper left', fontsize=12)
    ax1.set_title('Vibration and Clipping During GPS Signal Loss (200s - 380s)', fontsize=16)

    # 클리핑 데이터 시각화 (Clip)
    ax2 = ax1.twinx()  # 공유 x축
    ax2.plot(data['timestamp'], data['Clip'], label='Clip', color='red', linestyle=':', linewidth=2)
    ax2.set_ylabel('Clip Value', fontsize=14)
    ax2.tick_params(axis='y', labelcolor='red')
    ax2.legend(loc='upper right', fontsize=12)

    # 그래프 출력
    fig.tight_layout()
    plt.grid()
    plt.show()

# 함수 호출: 진동 및 클리핑 데이터 시각화
plot_vibration_and_clipping(gps_anomalous_data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 진동과 클리핑 데이터 상관 관계 분석
vibration_clipping_data = gps_anomalous_data[['VibeX', 'VibeY', 'VibeZ', 'Clip']]
correlation_matrix = vibration_clipping_data.corr()

# 상관 행렬 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Vibration and Clipping')
plt.show()
```

```python
# 모터 출력과 진동 데이터 상관 관계 분석
motor_vibration_data = gps_anomalous_data[['motor_1', 'motor_2', 'motor_3', 'motor_4', 'VibeX', 'VibeY', 'VibeZ']]
correlation_matrix_motor_vibe = motor_vibration_data.corr()

# 상관 행렬 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_motor_vibe, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Motor Outputs and Vibration')
plt.show()
```
